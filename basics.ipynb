```python
import numpy as np
from torch.autograd import Variable
import torch
inputs = np.random.rand(30,1)*2.0  #independent variable (input data),
weights = np.random.rand(2,1) #coefficient (weight) of the linear regression model

# y = mx + c 
# coff--> weights[0] & weights[1] 
# np.random.randn(30, 1) * 0.05 --> random noice with mean 0 and std 0.05

y = inputs*weights[0] + weights[1] + np.random.randn(30, 1) * 0.05
```

```python
print(weights)
```

```python
W = Variable(torch.rand(1, 1), requires_grad=True)
b = Variable(torch.rand(1), requires_grad=True)
```

```python
print(b)
print(W)
```

```python
import torch

# Define input tensor x (3x2)
xinput = torch.tensor([[1, 2],
                  [3, 4],
                  [5, 6]], dtype=torch.float32)

# Define weight tensor W (2x1)
Weits = torch.tensor([[0.5],
                  [0.7]], dtype=torch.float32)

# Perform matrix multiplication
result = torch.matmul(xinput, Weits)

print("Result of matrix multiplication:")
print(result)
```

```python
def linear(x):
    return torch.matmul(x, W) + b
```

```python
Xt = Variable(torch.from_numpy(inputs)).float()

yt = Variable(torch.from_numpy(y)).float()

for epoch in range(2500):

    # Compute predictions
    y_pred = linear(Xt)

    # Compute cost function
    loss = torch.mean((y_pred - yt) ** 2)

    # Run back-propagation
    loss.backward()

    # Update variables
    W.data = W.data - 0.005*W.grad.data
    b.data = b.data - 0.005*b.grad.data

    # Reset gradients
    W.grad.data.zero_()
    b.grad.data.zero_()
```

```python
import torch
import torch.nn.functional as F

class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)
        self.predict = torch.nn.Linear(n_hidden, n_output)

    def forward(self, x):
        x = F.relu(self.hidden(x))
        x = self.predict(x)
        return x

# Create an instance of the Net class
net = Net(n_feature=2, n_hidden=3, n_output=1)
```

* model fitting

```python
class Net(torch.nn.Module):
    def __init__(self, n_feature, n_hidden, n_output):
        super(Net, self).__init__()
        self.hidden = torch.nn.Linear(n_feature, n_hidden)   # hidden layer
        self.predict = torch.nn.Linear(n_hidden, n_output)   # output layer

    def forward(self, x):
        x = F.relu(self.hidden(x))      # activation function for hidden layer
        x = self.predict(x)             # linear output
        return x
```

```python
net = Net(n_feature=1, n_hidden=10, n_output=1) #definr the network
print(net)
```

* Optimizer - It is the most popular Optimization algorithms used in optimizing a Neural Network. Now gradient descent is majorly used to do Weights updates in a Neural Network Model , i.e update and tune the Model's parameters in a direction so that we can minimize the Loss function

```python
optimizer = torch.optim.SGD(net.parameters(), lr=0.2)
loss_func = torch.nn.MSELoss()  # this is for regression mean squared loss
```

```python
plt.ion()   # something about plotting
```

```python
for t in range(100):
    prediction = net(x)     # input x and predict based on x

    loss = loss_func(prediction, y)     # must be (1. nn output, 2. target)

    optimizer.zero_grad()   # clear gradients for next train
    loss.backward()         # backpropagation, compute gradients
    optimizer.step()        # apply gradients
    if t % 10 == 0:
        # plot and show learning process
        plt.figure(figsize=(20,6))
        plt.cla()
        plt.scatter(x.data.numpy(), y.data.numpy(), color = "orange")
        plt.plot(x.data.numpy(), prediction.data.numpy(), 'g-', lw=3)
        plt.text(0.3, 0, 'Loss=%.4f' % loss.data.numpy(), fontdict={'size': 25, 'color':  'red'})
        plt.show()
        plt.pause(0.1)

plt.ioff()
```
